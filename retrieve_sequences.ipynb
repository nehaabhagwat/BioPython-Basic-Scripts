```
from Bio import SeqIO
from Bio import Entrez
import pandas as pd

Entrez.email = "Enter email here"
```

```
temp_list = ['NM_000066.3', 'NM_178169.3', 'NM_178858.4']
for target in temp_list:
  print('Retrieving sequence for target transcript ' + target + '... ')
  handle = Entrez.efetch(db = 'nucleotide', id = target, rettype = 'gb', retmode = 'text')
  record = SeqIO.read(handle, 'genbank')
  print(str(record.seq))
  handle.close()
  
```
